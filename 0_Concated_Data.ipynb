```python
import pandas as pd

# Danh sách các tệp CSV cần gộp
file_paths = ['D:/DAP391m/data/house30-50.csv', 'D:/DAP391m/data/house50-60.csv', 'D:/DAP391m/data/house60-80.csv']

# Khai báo một DataFrame trống để lưu trữ dữ liệu từ các tệp CSV
merged_data = pd.DataFrame()

# Đọc từng tệp CSV và thêm dữ liệu vào DataFrame gộp
for file_path in file_paths:
    data = pd.read_csv(file_path)
    merged_data = pd.concat([merged_data, data], ignore_index=True)

# Lưu DataFrame gộp thành một tệp CSV mới
merged_data.to_csv('merged_file.csv', index=False)
```
