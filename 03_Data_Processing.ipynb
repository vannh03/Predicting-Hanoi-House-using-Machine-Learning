```python
def replace_missings (X) :
    X=X.drop(X.isna().sum()[X.columns[X.isna().sum() > 1000]].index, axis = 1)
    lis = X.isna().sum()[X.columns[X.isna().sum() > 1]].index
    for k in lis : 
        if type(X[k].iloc[0]) == str:
            X[k]=  X[k].fillna(X[k].iloc[0], inplace=True)
        
        else :
#             knn_imputerr=KNNImputer()
#             X[k]=knn_imputerr.fit_transform(X[k].values.reshape(-1, 1))
#             simple_imputer_mean = SimpleImputer(strategy='mean')
            X[k]=  X[k].round(fillna(X[k].mean()))
    return X 

def labelencoding (X) :
    lisle = []
    for col in X.columns:
        if X[col].dtype == 'object':
            label_encoder = LabelEncoder()
            X[col] = label_encoder.fit_transform(X[col])
            lisle.append(label_encoder)
    return X

def remove_outliers(X, threshold):
    z_scores = (X - X.mean()) / X.std()
    outliers = (z_scores.abs() > threshold).any(axis=1)
    X = X[~outliers]
    return X
```

```python
datatrain = pd.read_csv('./hanoi_house_final_4.csv')
datatrain= labelencoding(datatrain)
datatrain
datatrain = remove_outliers(datatrain, 5)
datatrain
```

```python
datatrain= replace_missings (datatrain)
```
